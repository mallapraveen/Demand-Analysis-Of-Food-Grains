```python
import pandas as pd

import folium
from folium import Choropleth, Circle, Marker
```

```python
df = pd.read_csv('../input/Current_Daily_Price_Of_Various_Commodities_From_Various_Markets_(Mandi)_With_Lat_Long.csv')
df.head(2)
```

```python
# Create a map
m_1 = folium.Map(location=[df.loc[0,'Latitude'], df.loc[0,'Longitude']], tiles='cartodbpositron', zoom_start=10)
m_1

# Add points to the map
for idx, row in df.iterrows():
    Marker([row['Latitude'], row['Longitude']]).add_to(m_1)

m_1
```
